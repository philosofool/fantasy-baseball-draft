```python
def filter_hitters(df, name=None, pos=None, availability=None):
    df = df.copy()
    if availability is None:
        availability = '^W $'
    df = df[df.Avail.str.contains(availability)]
    if name is not None:
        df = df[df.Player.str.contains(name, re.IGNORECASE)]
    if pos is not None:
        df = df[df.Eligible.str.match(f'{pos},|{pos}$')]
    return df.sort_values('fwar', ascending=False)

def filter_pitchers(df, has_saves=False, name=None, availability=None):
    df = df.copy()
    if availability is None:
        availability = '^W $'
    df = df[df.Avail.str.contains(availability)]
    if has_saves:
        df = df[df.S >= int(has_saves)]
    if name:
        df = df[df.Player.str.contains(name)]
    return df.sort_values('fwar', ascending=False)
```

```python
def update_avail(df, update: pd.DataFrame):
    order = df.columns
    out = df.drop(columns='Avail').merge(update[['Avail', 'Player']], how='left', on='Player')
    return out
```
